```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
print(tf.__version__)
```

# Simple Neural Network Implementation

```python
model1 = tf.keras.Sequential([keras.layers.Dense(units = 1, input_shape=[1])])
model1.compile(optimizer = "sgd", loss = "mean_squared_error")
```

```python
x = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
y = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
model1.fit(x, y, epochs=1000)
```

```python
print(model1.predict([10.0]))
```

# Fashion MNIST dataset - Neural Network Implementation

```python
fashion_mnist_dataset = tf.keras.datasets.fashion_mnist
(train_imgs, train_labels), (test_imgs, test_labels) = fashion_mnist_dataset.load_data()
# set1, set2 = fashion_mnist_dataset.load_data()
# train_imgs, train_labels
```

```python
x = tf.Variable([8])
x.shape
```

```python
train_imgs = train_imgs / 255.0
test_imgs = test_imgs / 255.0
train_imgs
```

```python
model2 = tf.keras.models.Sequential([tf.keras.layers.Flatten(),
                                     tf.keras.layers.Dense(128, activation = tf.nn.relu),
                                     tf.keras.layers.Dense(10, activation = tf.nn.softmax)])
```

```python
# len(train_imgs), len(train_labels)
model2.compile(optimizer = tf.keras.optimizers.Adam(),
               loss = "sparse_categorical_crossentropy",
               metrics = ['accuracy'])
model2.fit(train_imgs, train_labels, epochs = 5)
```

```python
model2.evaluate(test_imgs, test_labels)
```

```python
classifications = model2.predict(test_imgs)
print(classifications[0])
```
